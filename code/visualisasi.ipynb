```python
import matplotlib.pyplot as plt

data = {
    "Qonita": {
        "Netral": 100,
        "Senang": 100,
        "Sedih": 100
    },
    "Ridho": {
        "Netral": 100,
        "Senang": 100,
        "Sedih": 100
    },
    "Azanzi": {
        "Netral": 100,
        "Senang": 100,
        "Sedih": 100
    }
}
```

```python
# Data untuk bar chart
total_per_individu = {nama: sum(ekspresi.values()) for nama, ekspresi in data.items()}

labels_individu = total_per_individu.keys()
sizes_individu = total_per_individu.values()
colors_individu = ['#ff9999','#66b3ff',"#64d380"]

plt.figure(figsize=(8, 6))
plt.bar(labels_individu, sizes_individu, color=colors_individu)

plt.title('Bar Plot Data Individu', fontsize=16)
plt.xlabel('Individu', fontsize=12)
plt.ylabel('Jumlah Gambar', fontsize=12)
plt.ylim(0, max(sizes_individu) + 50)

# Menambahkan label angka di atas setiap bar
for i, v in enumerate(sizes_individu):
    plt.text(i, v + 10, str(v), color='black', ha='center', fontweight='bold')

plt.savefig('bar_chart_individu.png')
print("Bar chart untuk distribusi individu telah disimpan sebagai 'bar_chart_individu.png'")
```

```python
# Data untuk pie chart
total_per_ekspresi = {}
for subjek in data.values():
    for ekspresi, jumlah in subjek.items():
        if ekspresi not in total_per_ekspresi:
            total_per_ekspresi[ekspresi] = 0
        total_per_ekspresi[ekspresi] += jumlah

labels_ekspresi = total_per_ekspresi.keys()
counts_ekspresi = total_per_ekspresi.values()
colors_ekspresi = ['#c2c2f0','#ffcc99','#99ff99']

plt.figure(figsize=(8, 8))
plt.pie(counts_ekspresi, labels=labels_ekspresi, colors=colors_ekspresi,
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.title('Pie Chart Data Ekspresi', fontsize=16)
plt.axis('equal')
plt.savefig('pie_chart_ekspresi.png')
print("Pie chart untuk distribusi ekspresi telah disimpan sebagai 'pie_chart_ekspresi.png'")
```
